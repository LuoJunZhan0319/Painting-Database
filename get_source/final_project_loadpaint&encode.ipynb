### 將圖片轉base64編碼，放回df中

```python
import pandas as pd
import numpy as np
df = pd.read_csv("clear_data.csv",index_col=0)
```

```python
df = df[0:2599]
```

```python
df["paint_url"]
```

```python
#!/usr/bin/env python3
from PIL import Image
import base64,io

def tobase64(img):
    return base64.b64encode(img).decode('ascii')
def imgResize(img, scale=1):
    height, width = img.size 
    return img.resize((int(height*scale), int(width*scale)), Image.BILINEAR)

for i in range(2599):
    #開啟檔案
    im = Image.open("./images/"+str(i)+".jpg")
    #檔案變更尺寸
    rs_im = imgResize(im, 0.3)
    #格式轉換
    imgByteArr = io.BytesIO()
    rs_im.save(imgByteArr, format='PNG')
    #編碼並採用ascii
    encode = tobase64(imgByteArr.getvalue())
    df["paint_url"][i] = encode
```
