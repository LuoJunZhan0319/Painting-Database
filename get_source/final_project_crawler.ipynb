### 抓取所有名畫的類別網址

```python
#建立適當的 HTTP 請求
#匯入requests library
import requests

#將訪問該網址的結果回傳給 r
r = requests.get('https://www.ss.net.tw/')
r
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
#設立正規表達式
import re
regux = re.compile("^https://.*-[0-9]*\.html")

# class為 python保留自，查詢該屬性要加_
urls = soup.find_all("a", class_="list-group-item", href = regux)
print(urls) #可以將每個畫風類別的Link取下來
```

```python
# 使用find_all 後返回的是一組list
type(urls) #bs4.element.ResultSet
```

```python
print(type(urls[1])) #class 'bs4.element.Tag'
print(urls[1])
```

```python
links = [] #連結列表
for url in urls:
    links.append(url.get("href"))
links #留下url的部分
```

```python
def get_nextpagelink(soup):
    page = []
    urls = soup.find_all("a",class_="dropdown-item")
    for url in urls:
        if url.get('href') not in page:
            page.append(url.get('href'))
    page.pop(0)
    
    return page
```

### 測試抓畫

```python
import requests
from bs4 import BeautifulSoup
http = requests.get('https://www.ss.net.tw/paint-147.html#title')
soup = BeautifulSoup(http.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
def get_thisPagePicLink(soup):
    Piclink = []
    urls = soup.find_all("div",class_="card")
    for url in urls:
        a_tag = url.select("a")
        for a in a_tag:
            if a.get("href") not in Piclink:
                Piclink.append(a.get("href"))
    return Piclink
```

```python
import requests
import time
from bs4 import BeautifulSoup

def catch_picInfo(links = []):
    paints_info_list = []
    count = 0
    #每幅畫的link
    for link in links:
        #進到link 爬蟲
        name = ""
        http = requests.get(link)
        soup = BeautifulSoup(http.text, 'html.parser')
        
        ##抓畫源link
        paint_pics = soup.find("div", class_="col-md-5 col-lg-6")
        paint_info_list = []
        for pic in paint_pics:
            img = pic.select("img")
            for src in img:
                name = src.get("alt")
                paint_info_list.append(src.get("alt"))  #paint_name
                paint_info_list.append(src.get("data-src")) #paint_scr
                
        ##抓畫的資訊
        paint_infos = soup.find_all("ul", class_="list-unstyled")
        for info in paint_infos:
            paint_info_tag = info.select("li")
            for paint_info in paint_info_tag:
                paint_info_list.append(
                    paint_info.text.replace(u'\u3000',u'').
                    replace('編  號','').replace('作  者','').
                    replace('原作尺寸','').replace('原作材質','').
                    replace('瀏覽人次','').replace('年  份','').
                    replace('年份','').replace('作者','').replace('編號',''))
        #paint_info_list->品名、url、編號、作者、年分、尺寸、材質、瀏覽人次
        paints_info_list.append(paint_info_list)
        print("畫作{}的資料已爬取成功".format(name))
        count += 1
        time.sleep(0.5)
    print("這一頁抓了{}幅畫".format(count))
    return paints_info_list
```

### 開始抓圖

```python
#建立適當的 HTTP 請求
#匯入requests library
import requests
from bs4 import BeautifulSoup

#將訪問該網址的結果回傳給 r
r = requests.get('https://www.ss.net.tw/')
soup = BeautifulSoup(r.text, 'html.parser')

#設立正規表達式
import re
regux = re.compile("^https://.*-[0-9]*\.html")

# class為 python保留自，查詢該屬性要加_
urls = soup.find_all("a", class_="list-group-item", href = regux)

#將每個畫風類別的Link取下來
links = [] #連結列表
for url in urls:
    links.append(url.get("href"))
```

```python
links
```

```python
import time

all_paint_info = [] #所有畫作資訊儲存在這!

for link in links:
    #依序請求各網頁
    r = requests.get(link)
    #用BeautifulSoup解析網頁
    soup = BeautifulSoup(r.text, 'html.parser')
    #取得該網頁畫作連結
    paint_links = get_thisPagePicLink(soup)
    #抓每幅幅畫的資料，並回傳2維list
    all_paint_info.extend(catch_picInfo(paint_links))
    print("該連結{}以抓取完畢".format(link))

    #判斷是否有分頁
    page_links = get_nextpagelink(soup)
    if len(page_links) != 0:
        for page_link in page_links:
            r = requests.get("https://www.ss.net.tw/" + page_link)
            #用BeautifulSoup解析網頁
            soup = BeautifulSoup(r.text, 'html.parser')
            #取得該網頁畫作連結
            paint_links = get_thisPagePicLink(soup)
            #抓每幅幅畫的資料，並回傳2維list
            all_paint_info.extend(catch_picInfo(paint_links))
            print("該分頁{}以抓取完畢".format(page_link))
    
```

```python
import numpy as np
all_paint_info = np.asarray(all_paint_info,dtype=object)
np.savetxt('myfile.csv', all_paint_info, fmt='%s',encoding="utf-8")
```
